# Grouping Occupation Categories

```python
#checking the list of values within the occupation column

census_03["Occupation"].unique().tolist()
```

```python
#checking for the percentage of each value in the Occupation column

census_03["Occupation"].value_counts(normalize=True)
```

```python
#Appending the occupations with Retired to a specific value called Retired and PHD students to University students

census_03.loc[census_03["Occupation"].str.contains("Retired"), "Occupation"] = "Retired"
census_03.loc[census_03["Occupation"].str.contains("PhD Student"), "Occupation"] = "University Student"
```

```python
#counting the values again 

census_03["Occupation"].value_counts()
```

```python
#categorising the different occupations using enumerate
#creating a seperate column for classifying the occupation

Retired = census_03["Occupation"].str.contains("Retired")
University_Student = census_03["Occupation"] == "University Student"
Student = census_03["Occupation"] == "Student"
Unemployed = census_03["Occupation"] == "Unemployed"
Child = census_03["Occupation"] == "Child"

Occupation_col = []
for i, Occupation_column in enumerate(census_03["Occupation"]):
    if Retired[i] == True:
        Occupation_col.append("Retired")
    elif University_Student[i] == True:
        Occupation_col.append("University Student")
    elif Student[i] == True:
        Occupation_col.append("Student")
    elif Unemployed[i] == True:
        Occupation_col.append("Unemployed")
    elif Child[i] == True:
        Occupation_col.append("Child")
    else:
        Occupation_col.append("Employed")
        
census_03["Occupation_column"] = Occupation_col #New column for occupation 
```

```python
#Identifying those who identify as Head and whose ages are less than 18, equal to 18 and not equal to 18

census_03.loc[(census_03['Relationship to Head of House'] == 'Head') & (census_03['Age'] < 18), 'Head of House'] = "No"
census_03.loc[(census_03['Relationship to Head of House'] == 'Head') & (census_03['Age'] >= 18), 'Head of House'] = "Yes"
census_03.loc[(census_03['Relationship to Head of House'] != 'Head'), 'Head of House'] = "N/A"
census_03
```

# Plotting a Population Pyramid

```python
#using the age range of five years age groups
#creating an empty list

census_agerange = []
for element in census_03["Age"]:
    if element <= 4:
        census_agerange.append("0-4")
    elif element >= 5 and element < 10:
        census_agerange.append("5-9")
    elif element >= 10 and element < 15:
        census_agerange.append("10-14")
    elif element >= 15 and element < 20:
        census_agerange.append("15-19")
    elif element >= 20 and element < 25:
        census_agerange.append("20-24")
    elif element >= 25 and element < 30:
        census_agerange.append("25-29")
    elif element >= 30 and element < 35:
        census_agerange.append("30-34")
    elif element >= 35 and element < 40:
        census_agerange.append("35-39")
    elif element >= 40 and element < 45:
        census_agerange.append("40-44")
    elif element >= 45 and element < 50:
        census_agerange.append("45-49")
    elif element >= 50 and element < 55:
        census_agerange.append("50-54")
    elif element >= 55 and element < 60:
        census_agerange.append("55-59")
    elif element >= 60 and element < 65:
        census_agerange.append("60-64")
    elif element >= 65 and element < 70:
        census_agerange.append("65-69")
    elif element >= 70 and element < 75:
        census_agerange.append("70-74")
    elif element >= 75 and element < 80:
        census_agerange.append("75-79")
    elif element >= 80 and element < 85:
        census_agerange.append("80-84")
    elif element >= 85 and element < 90:
        census_agerange.append("85-89")
    elif element >= 90 and element < 95:
        census_agerange.append("90-94")
    elif element >= 95 and element < 100:
        census_agerange.append("95-99")
    else:
        census_agerange.append("100+")

census_03["Age_range"] = census_agerange #Creating a new caategory called agerange to categorise the age groups
```
